```python
import re

text = "My ID is 4567 and you can reach me at +1-234-567-8901 or 9876543210."


numbers = re.findall(r'\d+', text)
print("All numbers:", numbers)  

phone_numbers = re.findall(r'(?:\+?\d{1,3}[-.\s]?)?(?:\(?\d{2,4}\)?[-.\s]?){2,3}\d{3,4}', text)
print("Phone numbers:", phone_numbers) 
```
